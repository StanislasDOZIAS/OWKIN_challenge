```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def h(p):
    if p < 0.5:
        return 1/(1-2*p) * np.log((1-p)/p)
    else:
        return 1/(2*p*(1-p)) 

apply_h = np.vectorize(h)
```

```python
eps = 1e-2
X = np.linspace(eps, 1-5*eps, 100)
y = apply_h(X)
```

```python
plt.plot(X,y)
```
